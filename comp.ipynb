```python
import altair as alt
```

```python
data_m = data.merge(data_no,left_on = 'Load PF', right_on ='Load PF')
```

```python
data_m.columns
```

```python
data_m = data_m.rename(columns={'Load PF':'load_PF'})
```

```python
last_val = data_m['load_PF'].iloc[-1]
last_val
```

```python
chart1 = alt.Chart(data_m).mark_line(color = 'blue').encode(
    alt.X('load_PF', title ='Load PF'),
    alt.Y('Bus Voltage_x',title = 'Bus Voltage')
).properties(width = 500)
chart1 = chart1 + chart1.mark_text(align='left', dx=15, dy=-1, color = 'blue').encode(text=alt.condition(alt.datum.load_PF == last_val,
                            alt.value('With the Shunt of 300 MVAR connected at Bus 205'),
                            alt.value('')))
```

```python
chart1
```

```python
chart2 = alt.Chart(data_m).mark_line(color = 'green').encode(
      alt.X('load_PF', title ='Load PF'),
    alt.Y('Bus Voltage_y',title = 'Bus Voltage')
).properties(width = 500)
chart2 = chart2 + chart2.mark_text(align='left', dx=15, dy=10 ,color = 'green').encode(text=alt.condition(alt.datum.load_PF == last_val,
                            alt.value('Without the Shunt at Bus 205'),
                            alt.value('')))
```

```python
def function_plot(y_shunt, y_no_shunt, title_val):
    chart1 = alt.Chart(data_m, title = 'PV Analysis Comparison').mark_line(color = 'blue').encode(
        alt.X('load_PF', title ='Load PF'),
        alt.Y(y_shunt,title = title_val)
            ).properties(width = 500)
    chart1 = chart1 + chart1.mark_text(align='left', dx=15, dy=-1, color = 'blue').encode(text=alt.condition(alt.datum.load_PF == last_val,
                            alt.value('With the Shunt of 300 MVAR connected at Bus 205'),
                            alt.value('')))

    chart2 = alt.Chart(data_m).mark_line(color = 'green').encode(
        alt.X('load_PF', title ='Load PF'),
        alt.Y(y_no_shunt,title = title_val)
            ).properties(width = 500)
    chart2 = chart2 + chart2.mark_text(align='left', dx=15, dy=10 ,color = 'green').encode(text=alt.condition(alt.datum.load_PF == last_val,
                            alt.value('Without the Shunt at Bus 205'),
                            alt.value('')))
    chart = chart1+chart2
    return chart
```

```python
function_plot('Bus Voltage_x','Bus Voltage_y','Bus Voltage')
```

```python
function_plot('Load Pmax_x','Load Pmax_y','Load Pmax')
```
